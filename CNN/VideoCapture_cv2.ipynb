```python
import cv2
import os
import matplotlib.pyplot as plt
```

```python
filepath = './Office-Parkour.mp4'

if os.path.isfile(filepath):
  cap = cv2.VideoCapture(filepath)
else:
  print('nnononoonono')
```

```python
cap
```

```python
cap.get(cv2.CAP_PROP_FRAME_WIDTH)
```

```python
cap.get(cv2.CAP_PROP_FRAME_HEIGHT)
```

```python
from google.colab.patches import cv2_imshow

while True:
  retval, frame = cap.read() #retval : 정상적으로 읽어왔는지(True/False), frame : 한 장의 읽어온 이미지
  if not(retval):
    print('dfdfdf')
    break

  cv2_imshow(frame)
  key = cv2.waitKey(33)  #33msec 동안 한 프레임을 보여준다

  if key == 27:
    break

```

```python
if cap.isOpened():
  print('열렸음')
  cap.release()  #영상파일 사용을 종료
```
